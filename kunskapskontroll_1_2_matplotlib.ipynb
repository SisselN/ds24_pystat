```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Plotting can be done in an explicit (object oriented) or a implicit (function oriented) way. The explicit way is recommended.
In the object oriented approach a figure is explicitly created and methods are called directly from the axes:
fig, ax = plt.subplots()
ax.plot(x, y)

The implicit way works on a "current" figure and assumes which axes to use:
plt.plot(x, y)

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: the object on where the plot is drawn.
Axes: The objects the figure consists of.
Axis: Axes in singular (one dimension), part of an axes object.
Artist: All objects in matplotlib are artists, meaning they are subclasses of the artist class.

### When plotting in Matplotlib, what is the expected input data type?

NumPy arrays.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
y = x**2
fig, ax = plt.subplots()
ax.plot(x, y, color="DarkRed")
ax.set(title="OO approach", xlabel="x-axis", ylabel="y-axis")
ax.set_facecolor("BlanchedAlmond")

plt.figure()
plt.plot(x, y, color="DarkRed")
plt.title("PyPlot approach")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.gca().set_facecolor("LavenderBlush")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle("A figure with two subplots")
ax1.scatter(random_data_x, random_data_y,)
ax1.plot(x, y, color="DarkRed")
ax1.set(title="Scatter plot of random values.", xlabel="x", ylabel="y")

ax2.bar(names, values)
ax2.set(title="Bar plot of some fruits", xlabel="Some fruits", ylabel="Unknown variable")
plt.setp(ax2.get_xticklabels(), rotation=45, ha="right")
plt.tight_layout()
```
